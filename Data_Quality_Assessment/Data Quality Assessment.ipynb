```python
import pandas as pd

# Load dataset
data = pd.read_csv("Data Quality Assessment.csv")

# Check missing values
missing = data.isnull().sum()
print("Missing Values per Column:\n", missing)

# Check duplicates
duplicates = data.duplicated().sum()
print("Number of duplicate rows:", duplicates)

# Example: Fix missing values
data.fillna({'Column1': 'Unknown', 'Column2': 0}, inplace=True)

# Export cleaned dataset
data.to_csv("data_quality_cleaned.csv", index=False)
```
