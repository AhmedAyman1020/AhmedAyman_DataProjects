```python
import pandas as pd

# Load dataset
customers = pd.read_csv("Customer Behavior Analysis.csv")

# Data Cleaning
customers.drop_duplicates(inplace=True)
customers.fillna({'TotalAmount': 0}, inplace=True)
customers['OrderDate'] = pd.to_datetime(customers['OrderDate'])

# EDA
customer_summary = customers.groupby('CustomerID').agg({
    'OrderID':'count',
    'TotalAmount':'sum'
}).rename(columns={'OrderID':'NumOrders', 'TotalAmount':'TotalSpent'}).sort_values('TotalSpent', ascending=False)

top_customers = customer_summary.head(10)

print("Customer Summary:\n", customer_summary)
print("Top 10 Customers:\n", top_customers)
```
