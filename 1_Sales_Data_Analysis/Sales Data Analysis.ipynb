```python
import pandas as pd

# Load dataset
sales = pd.read_csv("Sales Data Analysis.csv")

# Data Cleaning
sales.drop_duplicates(inplace=True)
sales.fillna({'Price': sales['Price'].mean(), 'Quantity': 0}, inplace=True)
sales['Date'] = pd.to_datetime(sales['Date'])

# EDA
monthly_sales = sales.groupby(sales['Date'].dt.to_period('M')).agg({'SalesAmount':'sum'})
top_products = sales.groupby('ProductID').agg({'SalesAmount':'sum'}).sort_values('SalesAmount', ascending=False)

print("Monthly Sales:\n", monthly_sales)
print("Top Products:\n", top_products)
```
